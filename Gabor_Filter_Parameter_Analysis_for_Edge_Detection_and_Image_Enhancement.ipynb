```python
# gamma = 1
filter4 = cv2.getGaborKernel((51,51), 10, 0, (np.pi)/8, 1, psi=0, ktype=cv2.CV_32F)
plt.imshow(filter4, cmap='gray')
```

```python
# theta = 0-pi/2
filter5 = cv2.getGaborKernel((51,51), 10, (np.pi)/2, (np.pi)/8, 1, psi=0, ktype=cv2.CV_32F)
plt.imshow(filter5, cmap='gray')
```

```python
# theta = 0-pi/6
filter6 = cv2.getGaborKernel((51,51), 10, (np.pi)/6, (np.pi)/8, 1, psi=0, ktype=cv2.CV_32F)
plt.imshow(filter6, cmap='gray')
```

```python
# sigma = 1
filter7 = cv2.getGaborKernel((51,51), 1, (np.pi)/6, (np.pi)/8, 0, psi=0, ktype=cv2.CV_32F)
plt.imshow(filter7, cmap='gray')
```

```python
filters= []
thetas = np.linspace(0, np.pi / 2, 3)
sigmas = [1, 10, 3]
lambdas = [1, 2, 3]
gammas = [0, 1, 2]

```

```python
fig, axs = plt.subplots(len(sigmas), len(thetas), figsize=(15,15))

for i, sigma in enumerate(sigmas):
    for j, theta in enumerate(thetas):
        for lamda in lambdas:
            for gamma in gammas:
                # Generate the Gabor filter
                kernel = cv2.getGaborKernel((51, 51), sigma, theta, lamda, gamma, psi=0, ktype=cv2.CV_32F)
                filters.append(kernel)

                # Plot the filter in the corresponding subplot
                axs[i, j].imshow(kernel, cmap='gray')
                axs[i, j].set_title(f'Sigma={sigma}, Theta={theta:.2f}, Lambda={lamda}, Gamma={gamma}')

# Adjust subplot spacing and labels
plt.subplots_adjust(wspace=0.2, hspace=0.6)
for ax in axs.flat:
    ax.label_outer()

# Show the plot
plt.show()
```

# Conclusion:

A set of Gabor filters are generated for the following parameters:


1.   If sigma Value is changed from 1 to 5 then number of peaks increase.

2.   If filter size is reduced from 101 to 51 and gamma value is changed from 0.1 to 0 then strips of filter become vertical and intensity along each strip is same. If gamma value is changed from 0.1 to 1, the shape of Gabor filter change from eliptic curve to circular. The effect of changing the shape of filter is to reduce the effect on top, bottom and corners of the filtered image is the reduction in intensity of this part. This leads to reduction in intensity on top, bottom, left, right and corners of the filtered image and above reduction is dependent on elipticity of the gabor filter.

3.   If theta value is changed from 0 to pi/2, orientation of the filter changes from vertical to horizontal. This is useful in detecting the edges in a particular direction.

4.   Gabor filter with a specific set of parameters is appried in the image brick, the parameter value are applied horizontal and vertical edges of the image.

5.   A set of Gabor filter is applied on the book shelf image. It is observed that if filtered image has blurry edges, sigma value can be reduced to sharpen the blurry edges.

6.   To highlight edges in several direction, bank of gabor filter can be applied and identified edges can be overlaped on the original image.
