```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import io
import imutils
```

# SIFT when ContrastThreshold is 0.1

```python
img = cv2.imread("/content/NewYorkSkyline.jpg")
img_copy = img.copy()
```

```python
cv2_imshow(img_copy)
```

```python
img.shape
```

```python
sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=0.1)
```

```python
#Applying SIFT on original image
keypoints_img, descriptor_img = sift.detectAndCompute(img,None)
img_kp = cv2.drawKeypoints(img, keypoints_img, img, color=(0, 255, 0))
```

```python
print("count of keypoint of original image :", len(keypoints_img))
print("count of descriptor of original image :", len(descriptor_img))
print("count of each descriptor of original image :", len(descriptor_img[0]))
```

```python
cv2_imshow(img_kp)
```

```python
croped_img = img_copy[0:150,175:325]
cv2_imshow(croped_img)
```

```python
keypoints_croped_img, descriptor_croped_img = sift.detectAndCompute(croped_img,None)
croped_img_kp = cv2.drawKeypoints(croped_img, keypoints_croped_img, croped_img, color=(0, 255, 0))
```

```python
print("count of keypoint of croped image :", len(keypoints_croped_img))
print("count of descriptor of croped image :", len(descriptor_croped_img))
print("count of each descriptor of croped image :", len(descriptor_croped_img[0]))
```

```python
cv2_imshow(croped_img_kp)
```

```python
angle = 90
rotated_image = imutils.rotate(croped_img, angle)
cv2_imshow(rotated_image)

```

```python
keypoints_rotated_img, descriptor_rotated_img = sift.detectAndCompute(rotated_image,None)
rotated_img_kp = cv2.drawKeypoints(rotated_image, keypoints_rotated_img, rotated_image, color=(0, 255, 0))
```

```python
print("count of keypoint of rotated image :", len(keypoints_rotated_img))
print("count of descriptor of rotated image :", len(descriptor_rotated_img))
print("count of each descriptor of rotated image :", len(descriptor_rotated_img[0]))
```

```python
cv2_imshow(rotated_img_kp)
```

```python
len(keypoints_rotated_img)
```

```python
bf = cv2.BFMatcher()
```

```python
bf = cv2.BFMatcher()
matcher =bf.match(descriptor_img,descriptor_croped_img)
matcher_sorted = sorted(matcher, key=lambda val: val.distance)
img_kp_match= cv2.drawMatches(img, keypoints_img, croped_img, keypoints_croped_img, matcher_sorted[:10], None, matchColor=(0, 255, 0))
```

```python
len(matcher_sorted)
```

```python
cv2_imshow(img_kp_match)
```

```python
bf = cv2.BFMatcher()
matcher =bf.match(descriptor_img,descriptor_rotated_img)
matcher_sorted = sorted(matcher, key=lambda val: val.distance)
img_kp_match= cv2.drawMatches(img, keypoints_img, rotated_image, keypoints_rotated_img, matcher_sorted[:10], None , matchColor=(0, 255, 0))
cv2_imshow(img_kp_match)
```

# SIFT when Features are 100

```python
img = cv2.imread("/content/NewYorkSkyline.jpg")
img_copy = img.copy()
cv2_imshow(img_copy)
```

```python
img.shape
```

```python
sift = cv2.xfeatures2d.SIFT_create(nfeatures=100)
```

```python
#Applying SIFT on original image
keypoints_img, descriptor_img = sift.detectAndCompute(img,None)
img_kp = cv2.drawKeypoints(img, keypoints_img, img, color=(0, 255, 0))
```

```python
print("count of keypoint of original image :", len(keypoints_img))
print("count of descriptor of original image :", len(descriptor_img))
print("count of each descriptor of original image :", len(descriptor_img[0]))
cv2_imshow(img_copy)
```

```python
keypoints_croped_img, descriptor_croped_img = sift.detectAndCompute(croped_img,None)
croped_img_kp = cv2.drawKeypoints(croped_img, keypoints_croped_img, croped_img, color=(0, 255, 0))
print("count of keypoint of croped image :", len(keypoints_croped_img))
print("count of descriptor of croped image :", len(descriptor_croped_img))
print("count of each descriptor of croped image :", len(descriptor_croped_img[0]))
cv2_imshow(croped_img_kp)
```

```python
keypoints_rotated_img, descriptor_rotated_img = sift.detectAndCompute(rotated_image,None)
rotated_img_kp = cv2.drawKeypoints(rotated_image, keypoints_rotated_img, rotated_image, color=(0, 255, 0))
print("count of keypoint of rotated image :", len(keypoints_rotated_img))
print("count of descriptor of rotated image :", len(descriptor_rotated_img))
print("count of each descriptor of rotated image :", len(descriptor_rotated_img[0]))
cv2_imshow(rotated_img_kp)
```

```python
bf = cv2.BFMatcher()
matcher =bf.match(descriptor_img,descriptor_croped_img)
matcher_sorted = sorted(matcher, key=lambda val: val.distance)
img_kp_match= cv2.drawMatches(img, keypoints_img, croped_img, keypoints_croped_img, matcher_sorted[:10], None, matchColor=(0, 255, 0))
cv2_imshow(img_kp_match)
```

```python
bf = cv2.BFMatcher()
matcher =bf.match(descriptor_img,descriptor_rotated_img)
matcher_sorted = sorted(matcher, key=lambda val: val.distance)
img_kp_match= cv2.drawMatches(img, keypoints_img, rotated_image, keypoints_rotated_img, matcher_sorted[:10], None , matchColor=(0, 255, 0))
cv2_imshow(img_kp_match)
```

# Conclusion:


*   SIFT algorithm is applied on the original image to detect keypoints. For default values of SIFT, it shows 101 Keypoints and each keypoint has a corresponding descriptor and each descriptor has 128 elements.
*   Number of keypoints after extracting an object from the image is 27 and number of keypoints after rotating the extracting object is 18.

*   Brute Force matcher is used to match the keypoints of image1 and image2. Number of matches for original image and cropped image is 101. And number of matches for original image and rotated cropped image is 101.
*   Since the number of matches may include false keypoints, top 10 matches with minimum distances are shown in the image.

*   If contrast threshold is increases from default value of 0.3 to 0.1, we get 171 keypoints. The number of keypoints is lesser than the earlier case because contrast threshold has been increased to higher level which is done to filter out false keypoints.
